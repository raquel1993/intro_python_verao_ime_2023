```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/titanic.csv')
df
```

```python
#Questão 1
df['sobreviveu'].value_counts()
```

```python
#Questão 2
df.loc[(df['idade']) < 12].count()
```

```python
#Questão 3
df['embarque'].loc[(df['embarque'] == "Southampton") | (df['embarque'] == 'Queenstow')].count()
```

```python
#Questão 4
passageiras = df.loc[(df['sexo']) == 'feminino']
passageiras
```

```python
#Questão 5
criancas_cherbourg = df.loc[(df['idade']) < 12], [(df['embarque']) == 'Cherbourg']
criancas_cherbourg
```

```python
#Questão 6
titanic_menor = pd.DataFrame({'nome': df ['nome'],
                            'idade': df ['idade'],
                            'classe': df ['classe'],
                            'embarque': df ['embarque']})
titanic_menor
```

```python
#Questão 7
df2 = df.copy()
df2['tarifa_real'] = df2['tarifa']*6.43
df2
```

```python
#Questão 8
df.groupby(['sobreviveu']).count()
```

```python
#Questão 9
# Agrupe pelo local de embarque e calcule a média e a mediana da tarifa de cada grupo.
df[['embarque','tarifa']].groupby('embarque').agg(['mean', 'median'])
```

```python
#Questão 10
print("numero de pessoas por classe:")
df.groupby('classe')[['id_passageiro']].count()
```
